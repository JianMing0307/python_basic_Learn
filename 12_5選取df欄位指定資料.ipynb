# 選取df裡指定資料

### (1. Df[Bool Series]/2. Df.query('字串')/3. pd.concat())

```python
import numpy as np
import pandas as pd

scores = np.random.randint(50,101,size=(50,5))
scores_pd = pd.DataFrame(scores,columns=['國文', '英文', '數學', '地理', '歷史'],index=range(1,51))
scores_pd.index.name = "學號"
scores_pd.columns.name="科目"
scores_pd
```

## 1. df[Bool Series]

```python
bool_series= scores_pd["國文"] < 60
bool_series
```

```python
scores_pd[bool_series]
```

# 範例1

```python
import numpy as np
import pandas as pd

dataFrame = pd.read_csv('world.csv')
dataFrame
```

```python
dataFrame.info()
```

```python
dataFrame[dataFrame['國家'] == "台灣"]
```

```python
allDataFrame = dataFrame.reindex(columns=['洲名','國家','日期','總確診數','新增確診數','總人口數','解封指數'])

allDataFrame
```

```python
allDataFrame[allDataFrame['國家'] == '台灣']
```

## 2. df.query[ ] ------>選取較多欄位指定資料

```python
allDataFrame.query('國家=="台灣"')
```

```python
allDataFrame.query('國家=="台灣" or 國家=="日本"')
```

```python
allDataFrame.query('國家=="台灣" and 新增確診數==0')
```

```python
allDataFrame.query('國家=="台灣" and 新增確診數>=10000')
```

```python
allDataFrame.query('國家=="台灣"')['新增確診數'].max()
```

```python
taiwan_max = allDataFrame.query('國家=="台灣"')['新增確診數'].max()
taiwan_max_data = allDataFrame.query('國家=="台灣" and 新增確診數==@taiwan_max')       #注意語法:@變數
taiwan_max_data
```

```python
japan_max = allDataFrame.query('國家=="日本"')['新增確診數'].max()
japan_max_data = allDataFrame.query('國家=="日本" and 新增確診數==@japan_max')
japan_max_data
```

```python
us_max = allDataFrame.query('國家=="美國"')['新增確診數'].max()
us_max_data = allDataFrame.query('國家=="美國" and 新增確診數==@us_max')
us_max_data
```

```python
allDataFrame.query('國家=="台灣" and 日期>="2022-01-01" and 日期<="2022-01-03"')
```

## 3. df.concat()

```python
pd.concat([taiwan_max_data,japan_max_data,us_max_data])
```
