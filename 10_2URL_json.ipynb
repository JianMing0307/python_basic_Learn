# URL檔案讀寫

```python
from requests import get

url = 'https://raw.githubusercontent.com/roberthsu2003/python/master/%E6%AA%94%E6%A1%88%E5%AD%98%E5%8F%96/%E6%96%B0%E5%8C%97%E5%B8%82%E5%85%AC%E5%85%B1%E8%87%AA%E8%A1%8C%E8%BB%8A%E7%A7%9F%E8%B3%83%E7%B3%BB%E7%B5%B1.json'

response = get(url)
if response.ok:
    print(response.text)          #先將純文字檔從URL下載下來,並儲存
```

```python
#再將純文字檔轉成json檔
with open('新北市公共自行車租賃系統.json',mode='w',encoding='utf8') as file:
    file.write(response.text)
    print("存檔成功")
```

```python
import requests
url = 'https://tcgbusfs.blob.core.windows.net/dotapp/youbike/v2/youbike_immediate.json'

response = requests.get(url)
if response.ok:
    print(response.text)
```

```python
#純文字檔轉成python資料結構
import json

data_list = json.loads(response.text)
for item_dict in data_list:
    print(item_dict)
```

```python
ls = []
for item_dict in data_list:
    dic = {}
    dic["站點"] = item_dict["sna"][11:]
    dic["時間"] = item_dict["mday"]
    ls.append(dic)

ls
```

```python
#list comprehension
ls = [{"站點":item_dict["sna"][11:],"時間":item_dict["mday"]} for item_dict in data_list]
ls
```

```python
#python檔轉json
import json
with open("站點時間.json",mode="w",encoding="utf-8") as file:
    json.dump(ls,file,ensure_ascii=False)
print("存檔完成")
```

```python
#讀取json轉python 
import json

with open("站點時間.json",encoding="utf-8") as file:
    data_list = json.load(file)

data_list
```

```python
for item_dict in data_list:
    print(item_dict["時間"])   #輸出字串
    
```

```python
from datetime import datetime  #從datetime module裡面import datetime class
for item_dict in data_list:
    datetime_str = item_dict["時間"]
    current = datetime.strptime(datetime_str,"%Y-%m-%d %H:%M:%S")      #current是datetime實體    #strptime是datetime類別的method
    new_str = current.strftime("%m/%d/%Y %H/%M/%S")                                             #strftime是datetime/time/date的實體的method
    print(new_str)
```

```python
#使用timedelta取得一段的時間

from datetime import date, timedelta

now = date.today()
print(now)
one_day = timedelta(days=1)
tomorrow = now + one_day
print(tomorrow.year, tomorrow.month, tomorrow.day)

yesterday = now - one_day
print(yesterday.year, yesterday.month, yesterday.day)
```

# 利用comprehension快速建立資料結構

```python
#list append()
number_list = []
number_list.append(1)
number_list.append(2)
number_list.append(3)
number_list.append(4)
number_list.append(5)
number_list
```

```python
#list comprehension

number_list = [i for i in range(1,6)]
number_list
```

```python
#list comprehension
#2~100的偶數
number_list = [i for i in range(2,101) if i % 2 == 0]
number_list
```

```python
#dictionary comprehension
word = 'letters'
lettter_counds = {letter:word.count(letter) for letter in word}
lettter_counds
```
