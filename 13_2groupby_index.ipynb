```python
import numpy as np
import pandas as pd

url = 'https://media.taiwan.net.tw/od/01_PRD/%E6%AD%B7%E5%B9%B4%E5%9C%8B%E5%85%A7%E4%B8%BB%E8%A6%81%E8%A7%80%E5%85%89%E9%81%8A%E6%86%A9%E6%93%9A%E9%BB%9E%E9%81%8A%E5%AE%A2%E4%BA%BA%E6%95%B8%E6%9C%88%E5%88%A5%E7%B5%B1%E8%A8%88.csv'
dataFrame=pd.read_csv(url,encoding='utf-8')
dataFrame
```

```python
dataFrame1 = dataFrame.drop(columns=['類型','細分'])
dataFrame1
```

```python
dataFrame1.info()
```

```python
dataFrame2 = dataFrame1.dropna()
dataFrame2
```

```python
dataFrame2.info()
```

```python
dataFrame3 = dataFrame2.set_index(['年別','縣市別'])
dataFrame3
```

### (groupby( ))----->針對重複的index處理後 , 才能unstack( )

```python
dataFrame3.groupby(by=['縣市別']).sum(numeric_only=True)
```

```python
dataFrame4 = dataFrame3.groupby(by=['年別','縣市別']).sum(numeric_only=True)
             
dataFrame4
```

```python
dataFrame5 = dataFrame4.stack()
dataFrame5.index.names = ['年別', '縣市別', '月份']
dataFrame5
```

```python
dataFrame5.unstack('縣市別')
```

```python
dataFrame5.unstack('年別')
```

```python
dataFrame5.unstack('月份')
```

```python
dataFrame5.unstack(['縣市別','月份'])
```

# Example2

### (swaplevel( ))

```python
import numpy as np
import pandas as pd

dataFrame = pd.read_csv('各鄉鎮市區人口密度.csv')
dataFrame.columns = dataFrame.iloc[0]
dataFrame1 = dataFrame.drop(0)
dataFrame1
```

```python
#請取出 新北市中和區,新北市新莊區,嘉義縣水上鄉,臺東縣池上鄉,花蓮縣玉里鎮
#將這個整合為新的dataFrame
data1 = dataFrame1.query('區域別=="新北市中和區"')
data1
```

```python
data2 = dataFrame1.query('區域別=="新北市新莊區"')
data2
```

```python
data3 = dataFrame1.query('區域別=="嘉義縣水上鄉"')
data3
```

```python
data4 = dataFrame1.query('區域別=="臺東縣池上鄉"')
data4
```

```python
data5 = dataFrame1.query('區域別=="花蓮縣玉里鎮"')
data5
```

```python
all_data = pd.concat([data1, data2, data3, data4, data5])
all_data
```

```python
all_data1 = all_data.set_index(['統計年','區域別'])
all_data1
```

```python
all_data2 = all_data1.unstack('統計年')
all_data2
```

```python
all_data2.swaplevel(axis=1)
```

# Example3

```python
import numpy as np
import pandas as pd
df = pd.DataFrame({'key1':['a', 'a', 'b', 'b', 'a'],
                  'key2':['one', 'two', 'one', 'two', 'one'],
                  'data1': np.random.randn(5),
                  'data2':np.random.randn(5)})
df
```

### groupby(方法1)-columns先變index,在處理重複

```python
df1 = df.set_index(['key1','key2'])
df1
```

```python
df1.groupby(by=['key1', 'key2']).sum()
```

```python
df.groupby(df['key1']).sum()
```

### groupby(方法2)-colums直接轉index並處理重複

```python
df.groupby([df['key1'],df['key2']]).sum()
```
