```python
import matplotlib.pyplot as plt

index = ['apple', 'banana', 'orange', 'tomato', 'guava']
values = [5, 7, 3, 4, 6]

figure = plt.figure()
axes = figure.add_subplot(1,1,1)
axes.bar(index,values)

#調整xy軸座標位置
xmin, xmax, ymin, ymax = axes.axis()
print(xmin, xmax, ymin, ymax)
xmin -= 1
xmax += 1
ymax += 1
axes.axis(xmin=xmin, xmax=xmax, ymin=ymin, ymax=ymax)
axes.grid(axis='y')
plt.show()
```

```python
import matplotlib.pyplot as plt
# from matplotlib.font_manager import FontProperties
# font = FontProperties(fname='/Library/Fonts/Arial Unicode.ttf', size=8)
# font1 = FontProperties(fname='/Library/Fonts/Arial Unicode.ttf', size=10)
font = FontProperties(fname=r'c:\windows\fonts\simsun.ttc', size=8)
font1 = FontProperties(fname=r'c:\windows\fonts\simsun.ttc', size=10)

student1 = {'chinese':78, 'english':80, 'math':92, 'history':75, 'discover':85}
student2 = {'chinese':92, 'english':60, 'math':85, 'history':62, 'discover':73}
subjects = list(student1.keys())
scores1 = list(student1.values())
scores2 = list(student2.values())

figure = plt.figure(figsize=(8,5),dpi=220.0,facecolor='#cccccc')
axes = figure.add_subplot(1,1,1)
axes.set_title('學生1和學生2分數比較表',fontproperties=font)
axes.set_xlabel('科目',fontproperties=font)
axes.set_ylabel('分數',fontproperties=font)
axes.plot(subjects, scores1,'m-o')
axes.plot(subjects, scores2,'c-o')
axes.legend(['汪健明','張一正'],prop=font)
subjects_chinese = ['國文','英文','數學','歷史','科學']
axes.set_xticklabels(subjects_chinese,fontproperties=font)
xmin, xmax, ymin, ymax = axes.axis()
xmin += -0.2
xmax += 0.2
ymax = 100
axes.axis(xmin=xmin, xmax=xmax, ymin=ymin, ymax=ymax)

for index in range(len(subjects_chinese)):
    axes.annotate(str(scores1[index]),xy=(index,scores1[index]+1))
    axes.annotate(str(scores2[index]),xy=(index,scores2[index]+1))

score1_total = 0
for value in scores1:
    score1_total += value

score2_total = 0
for value in scores2:
    score2_total += value

axes.annotate(f"徐國堂總分{score1_total}\n張一正總分{score2_total}",xy=(2.7,90),fontproperties=font1)
plt.savefig('aa.png')
plt.show()
```

# 圓形圖表

```python
#圓形圖表

import matplotlib.pyplot as plt

labels = ['Nokia', 'Samsung', 'Apple', 'Lumia']
values = [20, 30, 45, 10]
colors = ['yellow', 'green', 'red', 'blue']

figure = plt.figure()
axes = figure.add_subplot()
axes.pie(values,labels=labels,colors=colors)
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['Nokia', 'Samsung', 'Apple', 'Lumia']
values = [20, 30, 45, 10]
colors = ['yellow', 'green', 'red', 'blue']

figure = plt.figure()
axes = figure.add_subplot()
axes.pie(values,labels=labels,colors=colors,explode=[0.3,0,0,0],shadow=True,autopct="%2.1f%%",startangle=180)
plt.show()
```
