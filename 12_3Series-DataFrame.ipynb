# 1. Series(一維)------->ndarray+index

### (建立Series)

```python
import pandas as pd

obj = pd.Series([4, 7, -5, 3])
obj
```

```python
obj1 = pd.Series([4, 7, -5, 3],index=['a','b','c','d'])
obj1
```

### (Series取值/改值)

```python
obj[0]
```

```python
obj[0] = 40
obj
```

### (陣列運算)

```python
stu1 = pd.Series([80, 90, 75, 65],index=['chinese', 'english', 'math', 'history'])
stu1
stu2 = pd.Series([67, 75, 73, 72],index=['chinese', 'english', 'math', 'history'])
stu2
```

```python
#所有學生+5分
stu2 += 5
stu2
```

```python
#Series + Series
total = stu1 + stu2
total
```

### (dict轉成Series)

```python
#dict變Series變bar圖

import matplotlib.pyplot as plt
sdata = {'Taipei':35000, 'Taoyuan':71000, 'Taichung':16000, 'Kaohsiung':5000}     #key當成Series的index,dict的value當成Series的value
obj3 = pd.Series(sdata)
index = obj3.index
values = obj3.values

figure = plt.figure(figsize=(8,5))
axes = figure.add_subplot()
axes.bar(index,values)   
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
sdata = {'Taipei':35000, 'Taoyuan':71000, 'Taichung':16000, 'Kaohsiung':5000}
obj3 = pd.Series(sdata)
index = obj3.index
values = obj3.values

plt.bar(index,values,width=0.8)
plt.show()
```

# 2. DataFrame(二維/n個Series)------->比Series多columns,

### (建立DataFrame)--->{key:list,key:list,key:list......}    Note:key當成columns

```python
import pandas as pd

data = {
    '縣市': ['台北','台北','台北', '新竹', '新竹', '新竹'],
    '年份': [2000, 2001, 2002, 2001, 2002, 2003],
    '人口': [2.2, 2.5, 2.7, 1.1, 1.5, 2.0]
}

frame1 = pd.DataFrame(data)
frame1
```

```python
#用np.random建立二維亂數,無法用python建立
import numpy as np
import pandas as pd

scores = np.random.randint(50,101,size=(50,5))   #50列5欄
print(scores)
scores_frame = pd.DataFrame(scores,
            columns=['國文','英文','數學','地理','歷史'],
            index=range(1,51)
            )
scores_frame.index.name="學號"
scores_frame.columns.name="科目"
scores_frame.name="學生分數表"
scores_frame
```

### (DataFrame取欄位)

```python
#取出國文成績----Series
scores_frame['國文'] 

#取出國文和數學成績----DataFrame
scores_frame[['國文','數學']]

sum_series = scores_frame['國文'] + scores_frame['數學'] + scores_frame['英文'] + scores_frame['地理'] + scores_frame['歷史']
scores_frame['總分'] = sum_series     #直接建立一個總分欄位
scores_frame
```

### (新增Df columns)

```python
#建立一個總分欄位
    #方法一
sum_series = scores_frame['國文'] + scores_frame['數學'] + scores_frame['英文'] + scores_frame['地理'] + scores_frame['歷史']
scores_frame['總分'] = sum_series     
scores_frame
    #方法二
scores_frame['總分'] = scores_frame.sum(axis=1)   #axis=1是整欄,axis=0是整列
scores_frame
```
