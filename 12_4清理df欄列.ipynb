## 範例1

```python
import pandas as pd

dataFrame = pd.read_csv('各鄉鎮市區人口密度.csv')
dataFrame
```

```python
dataFrame.columns = ['統計年', '區域別', '年底人口數', '土地面積', '人口密度']
dataFrame1 = dataFrame.drop(labels=0)
dataFrame1
```

### (檢查DataFrame資訊)

```python
#dataFrame.head()
#dataFrame.tail()
#dataFrame.info()
dataFrame1.head(n=10)
```

```python
dataFrame2 = dataFrame1.loc[:370]
```

```python
dataFrame2.info()    #所有都是字串
```

### (更改info內容:把數字str改成int)

```python
def parse_int(value):
    try:
        return int(value)
    except:
        return 0
    
dataFrame2['年底人口數'] = dataFrame2['年底人口數'].map(parse_int).copy()
```

```python
dataFrame2.info()
```

```python
dataFrame2['人口密度'] = dataFrame2['人口密度'].map(parse_int).copy()
```

```python
dataFrame2.info()
```

### (更改info內容:把數字str改成float)

```python
def parse_float(value):
    try:
        return float(value)
    except:
        return 0.0
    
dataFrame2['土地面積'] = dataFrame2['土地面積'].map(parse_float).copy()
```

```python
dataFrame2.info()
```

### (Df轉成excel)-灌excel viewer

```python
dataFrame2.to_excel('各鄉鎮市區人口密度.xlsx')
```

```python
import numpy as np
import pandas as pd

frame = pd.DataFrame(np.random.randn(4,3),
                    columns=list('bde'),
                    index=['台北','台中','台南','高雄'])
frame
```

```python
def apple_func(s):
    return s.max()

frame.apply(apple_func)
```

# 範例3

```python
import numpy as np
import pandas as pd

scores = np.random.randint(50,101,size=(50,5))
scores_pd = pd.DataFrame(scores,columns=['國文', '英文', '數學', '地理', '歷史'],index=range(1,51))
scores_pd.index.name = "學號"
scores_pd.columns.name="科目"
scores_pd
```

### (利用map更改index.name)

```python
scores_pd.index = scores_pd.index.map(lambda value:f'學號{value}')
scores_pd
```

```python
scores_pd.apply(lambda s:pd.Series([s.max(),s.min(),s.median(),s.std()],index=['最高分','最低分','中間分','標準差']))
```

```python
scores_pd.apply(lambda s:pd.Series([s.sum(),round(s.mean(),ndigits=2)],index=['總分','平均']),axis=1)
```

```python
class_dataFrame = scores_pd.apply(lambda s:pd.Series([s.sum(),round(s.mean(),ndigits=2)],index=['總分','平均']),axis=1)
```

### (排名次------->Series.rank())

```python
class_dataFrame['排名'] = class_dataFrame['平均'].rank(method='first',ascending=False)
class_dataFrame
```

### (排序------->df.sort_value())

```python
class_dataFrame.sort_values(by='排名')
```
